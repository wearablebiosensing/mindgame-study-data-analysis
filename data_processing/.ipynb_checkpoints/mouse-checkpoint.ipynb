```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import json
import re
import datetime
import seaborn as sns
```

# Extract User Distance Data

```python
# Directory where all of the data files are stored
DATA_DIRECTORY = "euclid/"
FILENAME_EXT = "json"

# Regex to extract data from filename
filename_pattern = re.compile(r"(shortest|user)_([NP]\d+)_L(\d+)_S(\d+)_([0-9]+)\.json")

# Prepare a list to store all data
data_columns = ["Type", "UserID", "Level", "Sublevel", "Timestamp", "Date", "Day", "Total_Distance", "json"]
data = []
shape_data = []

# Loop over each file in the directory
for filename in os.listdir(DATA_DIRECTORY):
    if filename.endswith(".json"):
        # Extract data from filename
        match = filename_pattern.search(filename)
        if match:
            datatype, user_id, level, sublevel, epoch = match.groups()
            date_obj = datetime.datetime.fromtimestamp(int(epoch))
            day = date_obj.strftime("%A")  # get full weekday name

            # Open the JSON file and read data
            with open(os.path.join(DATA_DIRECTORY, filename), "r") as file:
                json_data = json.load(file)
                
                # Calculate Total distance of all shapes in file
                total_distance = 0;
                for shape, distance in json_data.items():
                    total_distance += distance
                    

            # Prepare a row with extracted data
            row = [datatype, user_id, level, sublevel, epoch, date_obj, day, total_distance, json_data]
            data.append(row)
        else:
            print(f"Filename pattern did not match for file: {filename}")

# Create a DataFrame from the collected data
df = pd.DataFrame(data, columns=data_columns)

df.drop(columns="json").to_csv("mouse_data.csv")

# Display the DataFrame
df
```

# Per Shape

```python
# Filter the data for 'shortest' and 'user' types
shortest_df = df[df['Type'] == 'shortest']
user_df = df[df['Type'] == 'user']

shape_data = []
headers = ["user", "level", "sublevel", "timestamp", "shape", "distance", "percent_change"]

for index, row in user_df.iterrows():
    
    # Find the corresponding shortest distance row
    match = shortest_df[
        (shortest_df['UserID'] == row["UserID"]) &
        (shortest_df['Level'] == row['Level']) &
        (shortest_df['Sublevel'] == row['Sublevel']) &
        (shortest_df['Timestamp'] == row['Timestamp'])
    ]
    
    if not match.empty:
        for shape, distance in row["json"].items():
            # Convert the match JSON to a dictionary
            shortest_json = match["json"].iloc[0]
            percent_change = None
            
            if shape in shortest_json.keys():
                percent_change = abs(((distance - shortest_json[shape]) / shortest_json[shape]) * 100)
            # print(percent_change)
            
            # new_row = [row["UserID"],row['Level'], row['Sublevel'], row['Timestamp'], shape, distance,shortest_json.get(shape, "None"),  percent_change ]
            new_row = [row["UserID"],row['Level'], row['Sublevel'], row['Timestamp'], shape, distance,  percent_change ]
            
            shape_data.append(new_row)
            
            
shape_df = pd.DataFrame(shape_data, columns=headers)
shape_df.to_csv("mindgame_shapes.csv")

shape_df
        
```

# Percent Change & Acceleration

```python

# Filter the data for 'shortest' and 'user' types
shortest_df = df[df['Type'] == 'shortest']
user_df = df[df['Type'] == 'user']

# Convert timestamps to datetime for easier manipulation
# shortest_df['Timestamp'] = pd.to_datetime(shortest_df['Timestamp'], unit='s')
# user_df['Timestamp'] = pd.to_datetime(user_df['Timestamp'], unit='s')


level_df = pd.read_csv("mindgame_user_levels.csv")
level_df['date'] = pd.to_datetime(level_df['date'])
# print(level_df.dtypes, user_df.dtypes)

for index, row in user_df.iterrows():
    
    # Find the corresponding shortest distance row
    match = shortest_df[
        (shortest_df['UserID'] == row["UserID"]) &
        (shortest_df['Level'] == row['Level']) &
        (shortest_df['Sublevel'] == row['Sublevel']) &
        (shortest_df['Timestamp'] == row['Timestamp'])
    ]
    
    print(level_df[level_df["date"] == row["Date"]], row["Date"])
    level_match = level_df[
        (level_df['date'] == row["Date"]) 
        # (level_df['level'] == row['Level']) &
        # (level_df['sublevel'] == row['Sublevel']) 
    ]

    if not match.empty:
        shortest_distance = match.iloc[0]['Total_Distance']
        if shortest_distance > 0:
            percent_change = abs(((row['Total_Distance'] - shortest_distance) / shortest_distance) * 100)
            
            # Update the new column with the calculated percent change
            user_df.at[index, 'Percent_Change'] = percent_change
            # print(f"Recorded {percent_change}% change for user {row['UserID']}, level {row['Level']}")
            
        # print(level_match)
        if not level_match.empty :
            user_df.at[index, 'TimeToComplete(s)'] = level_match.iloc[0]['ttc']
            
            
        # Add acceleration data
        for filename in os.listdir("mouse_data/"):
            if filename.endswith(".csv"):
                filename_pattern = re.compile(r"([NP]\d+)_L(\d+)_S(\d+)_([0-9]+)\.csv")
                match = filename_pattern.search(filename)
                if match:
                    user_id, level, sublevel, epoch = match.groups()
                    if row['Timestamp'] == epoch:
                        path = os.path.join("mouse_data/", filename)
                        acc_df = pd.read_csv(path)
                        acc_df['x(px/s^2)'] = pd.to_numeric(acc_df['x(px/s^2)'], errors='coerce')
                        acc_df['y(px/s^2)'] = pd.to_numeric(acc_df['y(px/s^2)'], errors='coerce')
                        acc_df['x(px/s^2)'] = acc_df['x(px/s^2)'].abs()
                        acc_df['y(px/s^2)'] = acc_df['y(px/s^2)'].abs()
                        user_df.at[index, 'Average_X_Acceleration'] = acc_df['x(px/s^2)'].mean()
                        user_df.at[index, 'Average_Y_Acceleration'] = acc_df['y(px/s^2)'].mean()
                        user_df.at[index, 'Max_X_Acceleration'] = acc_df['x(px/s^2)'].max()
                        user_df.at[index, 'Max_Y_Acceleration'] = acc_df['y(px/s^2)'].max()
                        
                        

user_df.drop(columns=["Total_Distance", "json"], inplace=True)
user_df.to_csv("mindgame_level_percentchange.csv")
user_df
```

# Percent change Per User Per Level

```python

# Filter the data for 'shortest' and 'user' types
shortest_df = df[df['Type'] == 'shortest']
user_df = df[df['Type'] == 'user']

# Convert timestamps to datetime for easier manipulation
shortest_df['Timestamp'] = pd.to_datetime(shortest_df['Timestamp'], unit='s')
user_df['Timestamp'] = pd.to_datetime(user_df['Timestamp'], unit='s')

# Initialize a list to store results
results_list = []

# Iterate over users and levels
for user in user_df['UserID'].unique():
    for level in user_df['Level'].unique():
        # Filter both DataFrames for current user and level
        shortest_sub_df = shortest_df[(shortest_df['UserID'] == user) & (shortest_df['Level'] == level)]
        user_sub_df = user_df[(user_df['UserID'] == user) & (user_df['Level'] == level)]

        # Matching exact timestamps between 'user' and 'shortest'
        percent_changes = []
        for index, user_row in user_sub_df.iterrows():
            matching_row = shortest_sub_df[shortest_sub_df['Timestamp'] == user_row['Timestamp']]
            if not matching_row.empty and matching_row.iloc[0]['Total_Distance'] != 0:


                
                # Calculate percent change
                percent_change = abs(((user_row['Total_Distance'] - matching_row.iloc[0]['Total_Distance']) / matching_row.iloc[0]['Total_Distance']) * 100)
                # print(f"{user_row['UserID']} -> L{user_row['Level']}-S{user_row['Sublevel']} User Distance: {user_row['Total_Distance']}, Shortest Distance: {matching_row.iloc[0]['Total_Distance']} ::: PC = {percent_change}")
                percent_changes.append(percent_change)

        # Calculate average and standard deviation of percent changes
        if percent_changes:
 
            avg_change = np.mean(percent_changes)
            std_dev = np.std(percent_changes)
            # Append results to the list
            results_list.append({
                'User': user,
                'Level': level,
                'Average Percent Change': avg_change,
                'Standard Deviation': std_dev
            })

# Create DataFrame from results list
results_df = pd.DataFrame(results_list)
print(results_df)

# Print or save results
#print(results_df)
# Optionally save the results to a new CSV file
results_df.to_csv('averaged_percent_changes.csv', index=False)
```

```python
import pandas as pd

# Load the data
data = df

# Ensure Timestamp is in datetime format for easier manipulation
data['Timestamp'] = pd.to_datetime(data['Timestamp'], unit='s')

# Sort data to ensure correct ordering for grouping
data_sorted = data.sort_values(by=['UserID', 'Timestamp'])

# Separate the data for 'user' and 'shortest'
user_data = data_sorted[data_sorted['Type'] == 'user']
# user_data = user_data.sort_values(by=['UserID', 'Timestamp'])

shortest_data = data_sorted[data_sorted['Type'] == 'shortest']

# List to store overall results
results = []

# Process each user individually
for user in user_data['UserID'].unique():
    user_sessions = user_data[user_data['UserID'] == user]
    session_count = 0
    session_distances = []
    last_level = 0

    # Iterate through each user row
    for _, row in user_sessions.iterrows():

        
        # Start a new session if we hit level 1 and just finished level 3, or if it's the first row for this user
        if int(row['Level']) == 1 and int(last_level) == 3:
            if session_distances and len(session_distances) == 3:
                session_count += 1
                avg_percent_change = sum(session_distances) / len(session_distances)
                results.append({'UserID': user, 'Session#': session_count, 'Averaged_Percent_Change': avg_percent_change, "Level": row["Level"] })
                print(f"Session {session_count} for user {user} completed with Avg. Percent Change: {avg_percent_change}")
            session_distances = []  # Reset for new session
            last_level = 0
            
        print(f" {row['UserID']} - {row['Level']}, { int(last_level) }")

        # Only process if the level is sequentially next
        if int(row['Level']) == int(last_level) + 1:
            
            # Find the corresponding shortest distance row
            match = shortest_data[
                (shortest_data['UserID'] == user) &
                (shortest_data['Level'] == row['Level']) &
                (shortest_data['Sublevel'] == row['Sublevel']) &
                (shortest_data['Timestamp'] == row['Timestamp'])
            ]

            if not match.empty:
                shortest_distance = match.iloc[0]['Total_Distance']
                if shortest_distance > 0:
                    percent_change = abs(((row['Total_Distance'] - shortest_distance) / shortest_distance) * 100)
                    session_distances.append(percent_change)
                    print(f"Recorded {percent_change}% change for user {user}, level {row['Level']}")

            last_level = row['Level']

    # Check for the last session in case there are no more rows to trigger a new session
    if session_distances and len(session_distances) == 3:
        session_count += 1
        avg_percent_change = sum(session_distances) / len(session_distances)
        results.append({'UserID': user, 'Session#': session_count, 'Averaged_Percent_Change': avg_percent_change})
        print(f"Final Session {session_count} for user {user} completed with Avg. Percent Change: {avg_percent_change}")

# Convert results to a DataFrame
results_df = pd.DataFrame(results)

# Save the results to a CSV file
results_df.to_csv('output.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('mouse_data/mouse_N1_L1_S1_1712759963.csv')

# Prepare colors for each shape, adding a default color for null or any unspecified shape
color_map = {
    'No Shape': 'gray',
    'Trapezoid': '#A36DBD',
    'Hexagon': '#81E5DB',
    'Square': '#7BBEFC',
    'Yellow Diamond': '#E5CF81',
    'Equilateral Triangle': '#FFCC4D',
    'Purple Diamond': '#CA6B6E',
    'Circle': "#F08C8C",
    'Right Triangle': "#968FFF",
    "Quarter Circle": "pink"
}



# Handle 'null' and unspecified shapes by checking if shape is not in color_map
data['color'] = data['shape'].apply(lambda x: color_map.get(x, 'gray'))
data.fillna("NOSHAPE", inplace=True)

# Start a new figure
fig, ax = plt.subplots()

# Initialize variables to store segments
segment_x, segment_y, segment_colors = [], [], []
last_shape = None
used_shapes = set()  # Track used shapes for legend

# Iterate over the dataframe
for index, row in data.iterrows():

    if row['shape'] == 'END_OF_STROKE' or (last_shape == "NOSHAPE" and row['shape'] != 'NOSHAPE'):
        # Plot current segment if there are points collected
        if segment_x and segment_y:
            ax.plot(segment_x, segment_y, color=segment_colors[0])  # Use the color of the first point in the segment
            used_shapes.add(segment_colors[0])  # Add color to set
        # Reset for next segment
        segment_x, segment_y, segment_colors = [], [], []
    
    # Append coordinates and color if it's not an END_OF_STROKE
    if row['shape'] != 'END_OF_STROKE':
        segment_x.append(row['x'])
        segment_y.append(row['y'])
        segment_colors.append(row['color'])
        last_shape = row['shape']

# Plot any remaining data points after the last stroke
if segment_x and segment_y:
    ax.plot(segment_x, segment_y, color=segment_colors[0])
    used_shapes.add(segment_colors[0])

# Invert the y-axis to match screen coordinates
ax.invert_yaxis()

# Create legend only for the used shapes
handles = [
    plt.Line2D([0], [0], marker='o', color=color, linestyle='', markersize=10, label=shape)
    for shape, color in color_map.items() if color in used_shapes
]
ax.legend(handles=handles, title="Shape", loc='center left', bbox_to_anchor=(1, 0.5))

ax.set_xlabel('X Coordinate')
ax.set_ylabel('Y Coordinate')
ax.set_title('Mouse Movement Plot')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import re

def plot_mouse_movements(data, user_id, level, sublevel):
    # Prepare colors for each shape, adding a default color for null or any unspecified shape
    color_map = {
        'No Shape': 'gray',
        'Trapezoid': '#A36DBD',
        'Hexagon': '#81E5DB',
        'Square': '#7BBEFC',
        'Yellow Diamond': '#E5CF81',
        'Equilateral Triangle': '#FFCC4D',
        'Purple Diamond': '#CA6B6E',
        'Circle': "#F08C8C",
        'Right Triangle': "#968FFF",
        "Quarter Circle": "pink"
    }

    data['color'] = data['shape'].apply(lambda x: color_map.get(x, 'gray'))
    data.fillna("NOSHAPE", inplace=True)

    fig, ax = plt.subplots()

    segment_x, segment_y, segment_colors = [], [], []
    last_shape = None
    used_shapes = set()  

    for index, row in data.iterrows():
        
        # Once we accumulated all data from one movement we plot it
        if row['shape'] == 'END_OF_STROKE' or (last_shape == "NOSHAPE" and row['shape'] != 'NOSHAPE'):
            if segment_x and segment_y:
                ax.plot(segment_x, segment_y, color=segment_colors[0])  
                used_shapes.add(segment_colors[0])  
            segment_x, segment_y, segment_colors = [], [], []
        
        # Keep adding the x and y values
        if row['shape'] != 'END_OF_STROKE':
            segment_x.append(row['x'])
            segment_y.append(row['y'])
            segment_colors.append(row['color'])
            last_shape = row['shape']

    if segment_x and segment_y:
        ax.plot(segment_x, segment_y, color=segment_colors[0])
        used_shapes.add(segment_colors[0])

    ax.invert_yaxis()
    handles = [
        plt.Line2D([0], [0], marker='o', color=color, linestyle='', markersize=10, label=shape)
        for shape, color in color_map.items() if color in used_shapes
    ]
    ax.legend(handles=handles, title="Shape", loc='center left', bbox_to_anchor=(1, 0.5))

    ax.set_xlabel('X Coordinate')
    ax.set_ylabel('Y Coordinate')
    ax.set_title(f'User: {user_id}, Level: {level}, Sublevel: {sublevel}')

    return ax
```

```python
import os
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Directory and file handling
DATA_DIRECTORY = "mouse_data/"
FILENAME_EXT = "csv"
filename_pattern = re.compile(r"([NP]\d+)_L(\d+)_S(\d+)_([0-9]+)\.csv")

# Function to classify user type
def classify_user(user_id):
    if user_id.startswith('N'):
        return 'Neurotypical'
    elif user_id.startswith('P'):
        return 'Neurodivergent'
    else:
        return 'Unknown'

# Dictionary to store data for each user
user_data = {}

# Read and process each file
for filename in os.listdir(DATA_DIRECTORY):
    if filename.endswith(FILENAME_EXT):
        match = filename_pattern.search(filename)
        if match:
            user_id, level, sublevel, epoch = match.groups()
            path = os.path.join(DATA_DIRECTORY, filename)
            df = pd.read_csv(path)
            
            # Ensure numerical columns are correctly formatted
            df['x(px/s^2)'] = pd.to_numeric(df['x(px/s^2)'], errors='coerce')
            df['y(px/s^2)'] = pd.to_numeric(df['y(px/s^2)'], errors='coerce')
            
            if user_id not in user_data:
                user_data[user_id] = df
            else:
                user_data[user_id] = pd.concat([user_data[user_id], df])

# Calculate metrics for each user
user_metrics = {}



for user_id, data in user_data.items():
    data['x(px/s^2)'] = data['x(px/s^2)'].abs()
    data['y(px/s^2)'] = data['y(px/s^2)'].abs()
    metrics = {
        'mean_x_accel': data['x(px/s^2)'].mean(),
        'std_x_accel': data['x(px/s^2)'].std(),
        'max_x_accel': data['x(px/s^2)'].max(),
        # 'min_x_accel': data['x(px/s^2)'].min(),
        'mean_y_accel': data['y(px/s^2)'].mean(),
        'std_y_accel': data['y(px/s^2)'].std(),
        'max_y_accel': data['y(px/s^2)'].max(),
        # 'min_y_accel': data['y(px/s^2)'].min()
    }
    user_metrics[user_id] = metrics

# Convert metrics to DataFrame
metrics_df = pd.DataFrame.from_dict(user_metrics, orient='index')
# metrics_df.abs()

# Add user type to DataFrame
metrics_df['user_type'] = [classify_user(user_id) for user_id in metrics_df.index]

# Split data by user type
neurotypical_df = metrics_df[metrics_df['user_type'] == 'Neurotypical']
neurodivergent_df = metrics_df[metrics_df['user_type'] == 'Neurodivergent']

metrics_df



# Save aggregated data to CSV
neurotypical_df.to_csv('neurotypical_acceleration.csv', index=True)
neurodivergent_df.to_csv('neurodivergent_acceleration.csv', index=True)

# Plot the data
plt.figure(figsize=(16, 8))
sns.set(style="whitegrid")

# Box plots for mean accelerations
plt.subplot(2, 2, 1)
sns.boxplot(x='user_type', y='mean_x_accel', data=metrics_df)
plt.title('Mean X Acceleration')

plt.subplot(2, 2, 2)
sns.boxplot(x='user_type', y='mean_y_accel', data=metrics_df)
plt.title('Mean Y Acceleration')

# Scatter plots for mean accelerations
plt.subplot(2, 2, 3)
sns.barplot(x='user_type', y='std_y_accel', data=metrics_df, ci=None)
plt.title('Standard Deviation of Y Acceleration')

# Bar plots for standard deviations
plt.subplot(2, 2, 4)
sns.barplot(x='user_type', y='std_x_accel', data=metrics_df, ci=None)
plt.title('Standard Deviation of X Acceleration')

plt.tight_layout()
plt.show()

# Pair plot for in-depth analysis
sns.pairplot(metrics_df, hue='user_type', diag_kind='kde', markers=["o", "s"])
plt.suptitle('Pair Plot of Acceleration Metrics', y=1.02)
plt.show()
```
